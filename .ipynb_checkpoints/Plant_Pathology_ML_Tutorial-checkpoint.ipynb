## 1. A Prior on Images 
Before we dive into image classification, we first need to understand how an image is read on a computer. An image is a matrix of entries, also known as pixels, where each entry has a numerical value. Most often these values are represented by 8-bit integers (ranging from 0 to 255) that represent the intensity at that particular location. An image can be organized into multiple dimensions or "channels". For example, a Red-Green-Blue (RGB) image has 3 channels (Figure 1). The resulting color we see is the sum of the pixel values at that particular location.  <br>

## 2. Defining the Image Classification Problem 

Image classification aims to assign a label to an image. For example, recognize a "tulip" from a "rose" or a "cancerous region" from a "healthy region" of skin. The applications of image classification are literally infinite, but a strong image classification algorithm depends on good (and often copious amounts of) data. This is because a same object can be represented in an endless amount of settings. This then leads us to the challenges of image classification (Figure 2): 

<ol>
<li> <b>Viewpoint variation</b>. A single instance of an object can be oriented in many ways with respect to the camera.</li>
<li> <b>Scale variation</b>. Visual classes often exhibit variation in their size (size in the real world, not only in terms of their extent in the image).</li>
<li> <b>Deformation</b> Many objects of interest are not rigid bodies and can be deformed in extreme ways.</li>
<li> <b>Occlusion</b>. The objects of interest can be occluded. Sometimes only a small portion of an object (as little as few pixels) could be visible.</li>
<li> <b>Illumination conditions</b>. The effects of illumination are drastic on the pixel level.</li>
<li> <b>Background clutter</b>. The objects of interest may blend into their environment, making them hard to identify.</li>
<li><b>Intra-class variation</b>. The classes of interest can often be relatively broad, such as chair. There are many different types of these objects, each with their own appearance.</li>
</ol> 
    
    
A good image classification model must be invariant to the cross product of all these variations, while simultaneously retaining sensitivity to the inter-class variations.

### Build a Simple Image Descriptor


We will create a simple RGB histogram image descriptor. A color histogram can be used to count the number of times a specific pixel intensity occurs in an image. In this case, we will be doing this calculation for each channel in the image (Red-Green-Blue). So to put it simply, we will be calculating how many pixels of each color occur in a healthy plant image, and an unhealthy plant image. The counts for each color/color range will be used as our feature vector, which we will pass to our ML model. A graphical representation of a color histogram is given below.

Here, we will be constructing a 3D color histogram containing 8 bins per channel, which will produce feature vector of dims 8 x 8 x 8, or 512 when flattened. This is pretty handy, since the size of our original image is (800 x 800 pixels) x 3 channels or, 1 920 000 pixels. We are saving our computer a ton of memory space! If we had larger images (1920 x 1080 for example), it would be wise to reduce the image size to a more manageable dimension to increase the speed of the histogram computation.

```python
# Create a class for constructing our image descriptor. We will pass this descriptor to our classification model.

class RGBHist:
    def __init__(self, bins):
        # specify the number of bins that we will need to build the histogram
        self.bins = bins

    def describe(self, image):
        # compute a 3D histogram for each image using the 3 color spaces (R, G, B)
        # in the image. Normalize the histogram to accomodate multiple image
        # sizes if that is the case.
        histogram = cv2.calcHist([image], [0, 1, 2], None, self.bins, [0, 256, 0, 256, 0, 256])
        cv2.normalize(histogram, histogram)

        # return out 3D histogram as a flattened array
        return histogram.flatten()
```

```python
# initialize the image descriptor
desc = RGBHist([8, 8, 8])
```

Here we will define a function to create our training dataset. We will loop through the images, describe them using our image descriptor, and then pass the feature vectors to a Random Forest Classifier. Let's give this a go!

```python
# Function to help construct our training matrix
def create_training_data(trainingImgsPath, trainingLablsPath, descriptor):
    # create an empty list to store the images and labels
    trainingfeatures = []

    # sort the image paths so that they are matched with the correct label
    imagePaths = sorted(list(paths.list_images(trainingImgsPath)))

    # loop through the images and load them to disk
    for image in imagePaths:
        if "Test" in image: 
            pass
        else:
            img = cv2.imread(image)

            # convert the image from BGR to RGB format (opencv's default is BGR)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            if descriptor:
                # describe the image and update the list of features
                features = descriptor.describe(img)
                trainingfeatures.append(features)
            else: 
                # if there is no descriptor then return the loaded image only and resize the image to half width/height 
                img = resize(img, width=150)
                trainingfeatures.append(img)

    # load the training labels and only keep the 1st category
    trainingLabels = pd.read_csv(trainingLablsPath)
    trainingLabels = trainingLabels[["healthy"]].to_numpy().astype("uint8")

    return (trainingfeatures, trainingLabels)
```

```python
(X_train, y_train) = create_training_data(trainingImgsPath, trainingLablsPath, desc)
```

### Train the model

Next, we create a train/test split and train our model. We are putting aside a chunk of our data (25%) to use for evaluating its performance. 

```python
# construct the training and testing splits
(trainData, testData, trainTarget, testTarget) = train_test_split(X_train, y_train,
	test_size=0.25, shuffle = True)

print("Length of training data: ", len(trainData))
print("Length of validation data: ", len(testData))
```

```python
# train the classifier
RF_model = RandomForestClassifier(n_estimators=50, random_state=84)
RF_model.fit(trainData, trainTarget.ravel())
target_names = ['unhealthy', 'healthy']
```

### Evaluate the model 
The last step is checking to see how well our model did. We can use some handy functions from scikit-learn to do this pretty easily! 

```python
# evaluate the classifier
print(classification_report(testTarget, RF_model.predict(testData),
	target_names = target_names))
```

```python
# show the confusion matrix 
from sklearn.metrics import confusion_matrix
confusion_matrix(testTarget, RF_model.predict(testData))
```

```python
disp = plot_confusion_matrix(RF_model, testData, testTarget,
                                 display_labels=target_names,
                                 cmap=plt.cm.Blues, values_format =".0f")
disp.ax_.set_title("Confusion Matrix for the Random Forests Model")
```

Even with a simple RGB descriptor and Random Forest classifier, we were able to produce a model with an accuracy of 83%! Not bad. But maybe we can do better? Another very popular model that usually works well with images is the *Convolutional Neural Network*. Let's see if we can use it to improve our results.

### Using a Deep Learning (DL) Approach

#### Introduction to Convolutional Neural Networks 

Convolutional neural networks take advantage of the spatial patterns organized within an image. A CNN is composed of CONV and POOl layers primarily. The CONV layers learn specific filters that help extract key features without having to design these features ourselves. The POOL layers perform dimensionality reduction on the image after every set of convolutions. In the context of image classification, our CNN may learn to detect features like edges, shapes, textures or patterns, or even higher-level features like spatial patterns in the higher layers of the network. The  very last layer is a classifier that uses these higher-level features to make predictions regarding the contents of the image.

In practice, CNNs give us two key benefits: local invariance and compositionality.

#### Train the CNN 

We will create a new training matrix using only the original images reducing the image size from 784x784 to 150x150 pixels. 

```python
(X_train, y_train) = create_training_data(trainingImgsPath, trainingLablsPath, descriptor=None)
```

```python
# construct the training and testing splits
(trainData, testData, trainTarget, testTarget) = train_test_split(X_train, y_train,
	test_size=0.25, shuffle = True)
```

#### Assign a New Training Size 

Since training a CNN usually takes up a lot of space in memory, we will reduce the size of our images to half of the original pixel size. We will feed the images directly to the CNN and see what quality of predictions we get. 

```python
new_img_dim = (150, 150, 3)
```

#### Build the CNN

```python
from keras import layers
from keras import models
from keras import optimizers

# Build the CNN 
CNN_model = models.Sequential()
CNN_model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=new_img_dim))
CNN_model.add(layers.MaxPooling2D((2, 2)))
CNN_model.add(layers.Conv2D(64, (3, 3), activation='relu'))
CNN_model.add(layers.MaxPooling2D((2, 2)))
CNN_model.add(layers.Flatten())
CNN_model.add(layers.Dense(128, activation ="relu"))
CNN_model.add(layers.Dense(1, activation="softmax"))
CNN_model.summary()
```

```python
# compile the model 
CNN_model .compile(optimizer="rmsprop",
              loss="binary_crossentropy",
             metrics =["accuracy"])
```

Next, we will reshape and preprocess our training data to be able to successfully pass it to the CNN:

```python
# create our training and validation set for the CNN 
train_images = np.array(trainData).reshape((len(trainData), new_img_dim[0], new_img_dim[1], new_img_dim[2])) 
test_images = np.array(testData).reshape((len(testData), new_img_dim[0], new_img_dim[1], new_img_dim[2]))
```

Standardization of a dataset is a common requirement for many machine learning estimators: they might behave badly if the
individual features do not more or less look like normally distributed data. Here we standardize the image matrix by 
dividing all pixel values by the max pixel value, and setting all values between 0 and 1. This process is also called *Min-Max scaling*. 

```python
train_images = train_images.astype('float32')/255
test_images = test_images.astype('float32')/255
```

```python
# Train the model 
history = CNN_model.fit(train_images, trainTarget, epochs=10,batch_size=64)
```

Finally, let's compare how the CNN performs on the test data to see if it has learned how to generalize better. 

```python
# evaluate the classifier
test_loss, test_acc = CNN_model.evaluate(test_images, testTarget)
print("Loss on the test set: ", test_loss)
print("Accuracy on the set: \n", test_acc)

print("Classification Report")
CNN_predictions = CNN_model.predict(test_images)
print(classification_report(testTarget, CNN_predictions,
	target_names = target_names))

confusion_matrix(testTarget, CNN_predictions)
```

From what we have observed above, we can see that training a CNN on this dataset will not be a trivial task. Some of the reasons for this could be: 

<ol>
<li> Our network may not be deep enough to learn features that could differentiate between the two classes. </li>
<li> The model doesn't seem to be learning, it is primarily choosing the majority class within our dataset. This is a common problem in machine learning. </li>
<li> More hyperparameter tuning or a change in architecture might be needed to come up with a better model. </li>
<li> We could be stopping training too early. </li>
<li> We might just not have enough data. </li>
    
Let's visualize the loss of the model during training to see if this could give us any more information on what happened: 

```python
# Visualize the loss during training 
history_dict = history.history

loss_values = history_dict["loss"]
accuracy = history_dict["accuracy"]
epochs = range(1, len(loss_values)+1) 

plt.plot(epochs, loss_values, 'bo', label="Training Loss")
plt.plot(epochs, accuracy, 'r', label="Accuracy")
plt.title("CNN training loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

How could we maybe improve our model classification?

<ol>
<li> Use data augmentation to generate more images to train with. </li>
<li> Maybe consider using a pre-trained network as a feature extractor and build a new classifier. </li>
<li> Add forms of regularization to the model such as dropout. </li>
<li> Maybe this model architecture is not the best for our given problem! </li>
